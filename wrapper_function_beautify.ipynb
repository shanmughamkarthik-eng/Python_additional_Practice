###The beautify Decorator

####The Goal:
    Write a decorator named beautify that prints a line of dashes --- before and after any function it wraps.

```python
def beautify(func):
    def wrapper():
        # 1. Print a line like "----------"
        print("--------------------------------")
        # 2. Call the original function
        func()
        # 3. Print the same line "----------"
        print("--------------------------------")
        pass
    return wrapper

# Apply your decorator here
def greet():
    print("\nWelcome to the Python App!\n")

greet()
```

```python
beautify(greet)()
```

```python
greet = beautify(greet)

greet()
```

```python
greet()
```
