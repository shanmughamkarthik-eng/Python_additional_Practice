add two numbers and return the result using lambda function

```python
a=int(input("enter the first number: "))
b=int(input("enter the second number: "))
add = lambda a,b:a+b
print(f"Addition of {a} and {b} is: {add(a,b)}")
```

### lambda with single arguement

square a number

square = lambda x : x**2

```python
print("Lambda with single argument")
num = int(input("enter the number: "))
square = lambda x : x**2
print(f"Square of {num} is: {square(num)}")
```

## lambda with multiple arguments

area = lambda lenght, width : lenght * width

```python
print("Lambda with multiple arguments")
print("calculate area of rectangle using lambda function:")
width = int(input("enter the width: "))
length = int(input("enter the length: "))
area = lambda lenght, width : lenght * width
print(f"Area of rectangle is: {area(length,width)}")
```
