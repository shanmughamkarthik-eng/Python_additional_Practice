```python
### numbers as input
#calculate mean, median, mode

def calculate_mean(data):
  total_sum = 0
  length = len(data)

  for number in data:
    total_sum += number
  
  mean = total_sum / length
  return mean  
###############################

def calculate_median(data):
  sorted_numbers = sorted(data)
  length = len(sorted_numbers)

  if length % 2 == 0:
    median_1 =(length // 2 -1)
    median_2 = (length // 2)
    median = (sorted_numbers[median_1] + sorted_numbers[median_2]) / 2
  else :
    median = sorted_numbers[length // 2]

  return median
#########################################

def calculate_mode(data):
  freq = {}
  for num in data:
      freq[num] = freq.get(num, 0) + 1

  max_freq = max(freq.values())
  mode = [num for num, freq in freq.items() if freq == max_freq]
  
  all_same = len(set(freq.values())) == 1
  
  if all_same:
    print("All the numbers are in equal frequency.")

  return mode
####################################################################

data = [1, 10, 22, 24, 25]
mean = calculate_mean(data)
print(f"Mean: {mean}")

median = calculate_median(data)
print(f"Median: {median}")

mode = calculate_mode(data)
print(f"Mode: {mode}")
```
