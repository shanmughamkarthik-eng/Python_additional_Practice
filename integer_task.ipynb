```python
num = int(input())

if 0 <= num <= 100:
  if num % 2 != 0:
    print("Weird")
  elif num % 2 == 0 and 2 <= num <= 5:
    print("Not Weird")
  elif num % 2 == 0 and 6 <= num <= 20:
    print("Weird")
  elif num % 2 == 0 and num > 20:
    print("Not Weird")

else:
  print("please enter the number within range of 0 and 100 again")
```
