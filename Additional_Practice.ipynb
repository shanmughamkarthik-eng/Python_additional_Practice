```python
# Compound Interest:
# Create a program that takes principal, rate, and time (float) and calculates the final amount: \(A=P(1+r/n)^{nt}\).
print("Create a program that takes principal, rate, and time (float) and calculates the final amount")
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the annual interest rate: "))
time = float(input("Enter the time in years: "))

amount = principal * (1 + rate / 100) ** time

print(f"The final amount is: {amount:.2f}")
```

```python
# Bank Withdrawals: Given a float balance, ask the user for a withdrawal_amount.
# If the amount is greater than the balance, print an error;
# otherwise, print the new balance rounded to 2 decimals.
print("Bank Withdrawals ")

balance = float(input("Enter the current balance: Rs."))
withdrawal_amount = float(input("Enter the withdrawal amount: Rs."))

if withdrawal_amount > balance:
    print("Error: Withdrawal amount exceeds the balance.")
else:
    new_balance = balance - withdrawal_amount
    print(f"The new balance is: Rs.{new_balance:.2f}")
```

```python
# Length: Write a program to calculate the length of a string
print("Length: Write a program to calculate the length of a string")

my_string = input("Enter a string: ")
length = len(my_string)

print(f"Length of string {my_string} is : {length}")
```

```python
# Reverse: Reverse a string using slicing.
print("Reverse: Reverse a string using slicing.")

my_string = input("Enter a string: ")
reversed_string = my_string[::-1]
print(reversed_string)
```

```python
# Case Swap: Convert all uppercase letters to lowercase and vice versa
print("Case Swap: Convert all uppercase letters to lowercase and vice versa")

my_string = input("Enter a string: ")
swapped_string = my_string.swapcase()
print(my_string)
print(swapped_string)
```
